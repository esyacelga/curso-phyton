Regresion con arboles de desicion 

Importando el DATA SET

```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv("./Position_Salaries.csv")

dataset.head()
x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

Ajustar la regresion con el data set

```python
from sklearn.tree import DecisionTreeRegressor
regression = DecisionTreeRegressor(random_state=0)
regression.fit(x,y)
```

Prediccion de nuestros modelos

```python
y_pred = regression.predict([[6.5]])
print(y_pred)
```

Visualizacion de los resultados del modelo polinomico


```python

x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x, y, color="red")
plt.plot(x, regression.predict(x), color="blue")
plt.title("Sueldo vs Años de experiencia (Conjunto de datos de TESTING)")
plt.xlabel("Años de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```
