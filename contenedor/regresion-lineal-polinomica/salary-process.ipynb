1. Importacion de datos

```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv("./Position_Salaries.csv")

x = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values

print (y)
```

2. Divicion del dataset en entrenamiento y test

```python
#from sklearn.model_selection import train_test_split
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

3. Ajustar la regresion lineal con el Dataset

```python
from sklearn.linear_model import LinearRegression
regresion = LinearRegression();
regresion.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 =  LinearRegression()
lin_reg_2.fit(x_poly,y)
```

4 Visualizacion de los resultados del modelo polinomico


```python
plt.scatter(x,y, color="red")
plt.plot(x , regresion.predict(x), color="blue")
plt.title ("Modelo de regresion lineal")
plt.xlabel("Posicion del empleado")
plt.ylabel("Sueldo en dolares")
#plt.show()
```

```python
import numpy as np

x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)

plt.scatter(x,y, color="red")
plt.plot(x , lin_reg_2.predict(x_poly), color="blue")
plt.title ("Modelo de regresion Polinomica")
plt.xlabel("Posicion del empleado")
plt.ylabel("Sueldo en dolares")



```

```python
plt.scatter(x,y, color="red")
plt.plot(x_grid , lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color="blue")
plt.title ("Modelo de regresion Polinomica")
plt.xlabel("Posicion del empleado")
plt.ylabel("Sueldo en dolares")
```

5. Prediccion de nuestros modelos

```python
regresion.predict([[6.5]])
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
